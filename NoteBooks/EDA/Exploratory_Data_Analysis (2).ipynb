```python
#Which drug categories contribute the most to revenue?

df.groupby('drugcategory')['totalsales'].sum().sort_values(ascending=False).plot(kind='bar', title='Revenue by Drug Category')
```

- Antidiabetic drugs contribute the most to total revenue, followed by Cardiac and Antibiotic categories.



```python
#Are we generating revenue through volume or value?
df['RevenuePerUnit'] = df['totalsales'] / df['unitssold']
df.groupby('drugcategory')[['unitssold', 'RevenuePerUnit']].mean().plot(kind='bar', subplots=True, layout=(1,2), figsize=(12,4), title=['Avg Units Sold', 'Avg Revenue Per Unit'])
```

- Antidiabetic drugs lead in both average units sold and revenue per unit, indicating strong performance in both volume and value.
- Analgesics sell in high volumes but have the lowest revenue per unit, suggesting they are volume-driven, low-margin products


```python
#Which products should be prioritized for growth?
monthly_product_sales = df.groupby(['drugname', df['date'].dt.to_period('M')])['totalsales'].sum().unstack().fillna(0)
growth_candidates = monthly_product_sales.pct_change(axis=1).mean(axis=1).sort_values(ascending=False).head(10)
print(growth_candidates)
```

- Atorvastatin is the top revenue contributor (~94%), making it a high-priority product for retention and growth.
- Metformin, despite being a common antidiabetic drug, contributes the least (~4%), suggesting low pricing or underperformance, and may need strategic review or repositioning.


```python
#Which regions and states are over- or under-performing?
region_perf = df.groupby('region')['totalsales'].sum().sort_values(ascending=False)
region_perf.plot(kind='bar', title='Total Sales by Region')
```

- The South region is the top performer, generating the highest total sales, followed by East and West.
- The North region lags significantly behind, indicating potential underperformance or untapped market opportunity.


```python
#Where does the company have untapped market potential?

df.groupby('region')['unitssold'].sum().div(df.groupby('region')['totalsales'].sum()).sort_values().plot(kind='bar', title='Units Sold per Revenue (Proxy for Market Efficiency)')
```

- The West region has the highest units sold per revenue, suggesting it operates in a high-volume, low-value market - a sign of potential for premium product penetration.
- The East region shows the lowest efficiency, indicating lower unit sales per revenue, possibly due to higher-priced drugs or underutilized sales capacity - candidate for volume expansion.


```python
#Are certain drugs region-dependent?
# Import seaborn for heatmap functionality
import seaborn as sns
import matplotlib.pyplot as plt

# Create the pivot table
pivot_table = pd.pivot_table(df, index='drugname', columns='region', values='totalsales', aggfunc='sum').fillna(0)

# Create a figure with appropriate size
plt.figure(figsize=(10, 8))

# Use seaborn's heatmap function
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt=".1f")

# Add a title
plt.title('Drug Sales by Region')

# Show the plot
plt.show()
```

- Insulin shows a strong regional skew, with the East region contributing the highest sales (~1.6M), indicating region-specific demand or prescribing patterns.
- Losartan and Atorvastatin are heavily sold in the South and West, while Amoxicillin dominates in the East, suggesting regional preferences or therapeutic needs.


```python
# Are some reps active but inefficient
rep_perf = df.groupby('salesrep').agg({'totalsales':'sum', 'invoiceid':'count'})
rep_perf['RevenuePerInvoice'] = rep_perf['totalsales'] / rep_perf['invoiceid']
sns.scatterplot(data=rep_perf, x='invoiceid', y='RevenuePerInvoice', hue='totalsales', size='totalsales')
```

- Bubbles with high invoice counts but low revenue per invoice (toward the bottom-right) indicate active yet inefficient reps, generating less value per transaction.
- In contrast, bubbles in the top-left with fewer invoices but high revenue per invoice suggest strategic, high-value selling, ideal for modeling or incentives.


```python
#Who should be considered for incentives or training?
rep_perf['Efficiency'] = pd.qcut(rep_perf['RevenuePerInvoice'], q=2, labels=['Low', 'High'])
rep_perf['Activity'] = pd.qcut(rep_perf['invoiceid'], q=2, labels=['Low', 'High'])
print(rep_perf[['Efficiency', 'Activity']])
```

- Rahul Das stands out with High Efficiency and High Activity, making him an ideal candidate for incentives or leadership modeling.
- Ankit Singh, Neha Shah, and Sneha Iyer fall into the Low-Low quadrant, indicating a need for training, support, or territory reassessment.

```python
#Are there any extreme or anomalous sales values that could distort analysis?

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Example: Detect outliers in TotalSales
outliers_sales = detect_outliers_iqr(df, 'totalsales')
print(f"Outliers in totalsales: {len(outliers_sales)} rows")
```

```python
df['MonthYear'] = df['date'].dt.to_period('M')
monthly_trend = df.groupby('MonthYear')['totalsales'].sum()
monthly_trend.plot(title='Monthly Revenue Trend', marker='o')
plt.ylabel('Revenue')
plt.show()
```

- This suggests strong seasonal demand mid-year, followed by a potential slowdown or market saturation post-August.

```python
volatility = df.groupby('MonthYear')['totalsales'].std()
volatility.plot(title='Monthly Revenue Volatility', color='orange')
plt.ylabel('Std Dev of Revenue')
plt.show()
```

- Peaks in September and December 2024 indicate periods of high variability, possibly due to promotions, seasonal demand, or inconsistent sales patterns.
- Troughs in October 2024 and February 2025 reflect more stable revenue, suggesting consistent performance or fewer market disruptions during those months.

## Key Insights

## Antidiabetic drugs are the primary revenue driver

Antidiabetic drugs contribute the highest total revenue and average units sold, outperforming all other categories. Their high revenue per unit and variability indicate premium pricing opportunities and strong market demand.

## Revenue is highly concentrated in a few products

Atorvastatin alone contributes ~94% of total revenue, making it a critical product for business sustainability. In contrast, Metformin underperforms despite high usage, signaling a need for pricing, positioning, or volume strategy review.

## Strong regional imbalance in sales performance

The South region leads overall revenue, followed by East and West, while the North region significantly lags, indicating untapped market potential or weaker sales execution in that region.

## Efficiency varies significantly across regions

The West region shows the highest units sold per revenue, suggesting a price-sensitive, volume-driven market with potential for premium product introduction. The East region shows lower efficiency, indicating scope for demand generation or better sales optimization.

## Sales performance depends on efficiency, not just volume

Among sales representatives, Ravi Kumar generates the highest revenue per invoice, showing superior sales efficiency, while Rahul Das leads in total revenue through high transaction volume. This highlights the importance of balancing productivity with effectiveness.
